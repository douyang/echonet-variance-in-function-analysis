```python
###### 
# For given set of MRN + Date and a set of ECGs (matched by MRN and with all dates)
# Identify and output closest ECG by date thats MRN matched
######

# David Ouyang 9/10/2020

import io
import numpy as np
import datetime

ECGsFileName = "C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNet_ECG.csv"

OutECGsFileName = "C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNet_ECG_Paired.csv"

HashedNameFile = "C:\\Datasets\\IdentifiedData\\NamesToHashedNames-Full-wMRN.csv"
HashedNameDict = {}
```

```python
with open(HashedNameFile, 'r') as Mapping:
    for line in Mapping.readlines():
        #rint(line)
        lineSplit = line.strip().split(',')
        hashedFileName = lineSplit[1]
        try:
            date = datetime.datetime.strptime(lineSplit[4],'%Y%m%d')
        except:
            date = lineSplit[4]
        MRN = lineSplit[5]
        
        HashedNameDict[hashedFileName] = (date, MRN)
```

```python
with open(OutECGsFileName, 'w') as outFile:
    for video in HashedNameDict.keys():
        (date,MRN) = HashedNameDict[video]
        
        #print(video, date, MRN)
        
    
        closest = "No Match"
        closestTimeDiff = 9999999
        
        if(MRN != 'NotFound'):
            with open(ECGsFileName, 'r') as ECGsFile:
                for line in ECGsFile.readlines():
                    #print(len(line.split(',')))
                    if len(line.split(',')) > 6:
                        try:
                            match = line.split(',')[1].strip()
                            if(match == MRN):
                                thisDate = datetime.datetime.strptime(line.split(',')[3], "%Y-%m-%d %H:%M:%S")#"%m/%d/%Y %H:%M")
                                thisDiff = abs((date - thisDate).days)

                                if thisDiff < closestTimeDiff:
                                    closestTimeDiff = thisDiff
                                    closest = line.split(',')[4] + "," + line.split(',')[5]

                                    #print(MRN, thisDiff)
                        except:
                            print("misaligned, potentially added comma with 'DOE,JOHN' in name")
                            
        elif(MRN == 'MRN'):
            #closest = "EchoDate,MRN,DaysOff,MRN_match,MRN_search,name,dateAcquired,patientId,ecgId,heartRate,prInterval,qtInterval,qrsDuration,pDuration,rrInterval,qtcb,pFrontAxis,qrsFrontAxis,stFrontAxis,tFrontAxis,pHorizAxis,qrsHorizAxis,stHorizAxis,tHorizAxis,pHorizAxis,measurmentInfo,interpretation\n"
            closest = "patientID,ecgID"
        if(MRN != 'NotFound'):
            try:
                date = date.date()
            except:
                date = date

                
            #print(video, date, MRN, closestTimeDiff, closest)

            outFile.write(video + "," +str( date ) + "," + MRN + "," + str(closestTimeDiff) + "," + closest  + "\n")

        
        
```

```python
patID = {}
ecgID = {}

with open(OutECGsFileName, 'r') as HashedNameToECGFile:
    for line in HashedNameToECGFile.readlines():
        lineSplit = line.strip().split(',')
        #print(lineSplit)
        filename = lineSplit[0]
        diff = int(lineSplit[3])
        
        if diff <= 30:
            thisPatID = lineSplit[4]
            thisEcgID = lineSplit[5]
            patID[thisPatID] = filename
            ecgID[thisEcgID] = filename
            
        #print(line)
```

```python
import os
from zipfile import ZipFile

with ZipFile("C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNetECGs\\EchoNet_ECG_waveforms.zip", 'r') as zipObject:
    listOfFileNames = zipObject.namelist()
    for i, fileName in enumerate(listOfFileNames):
        fileName = fileName.replace("EchoNet_ECG_waveforms/","").replace(".npy", "")

        if fileName in ecgID.keys():
            print(fileName, "in ecgID")
            zipObject.extract("EchoNet_ECG_waveforms/" + fileName + '.npy', "C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNetECGs")
            os.rename(os.path.join("C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNetECGs\\EchoNet_ECG_waveforms", fileName + '.npy'), os.path.join("C:\\Datasets\\IdentifiedData\\EchoNetECGs\\EchoNetECGs\\EchoNet_ECG_waveforms", ecgID[fileName] +  '.npy'))
        
```
